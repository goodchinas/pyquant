# pyecharts图样式详解

```python
import pyecharts as pe
import tushare as ts
import pandas as pd
```

# 条形图

```python
a=ts.get_gdp_quarter()
a.head()
```

返回值说明：

-  quarter :季度
-  gdp :国内生产总值(亿元)
-  gdp\_yoy :国内生产总值同比增长(%)
-  pi :第一产业增加值(亿元)
-  pi\_yoy:第一产业增加值同比增长(%)
-  si :第二产业增加值(亿元)
-  si\_yoy :第二产业增加值同比增长(%)
-  ti :第三产业增加值(亿元)
-  ti\_yoy :第三产业增加值同比增长(%)

```python
b=a.iloc[0,[3,5,7]]
b.head()
```

```python
bar=pe.Bar('最新季度GDP增加额','条形图')
bar.add('三产业增加额',b.keys(),b.values)
bar
```

# 水平条形图

```python
bar=pe.Bar(title='最新GDP增加额',subtitle='水平条形图')
bar.add('增加额',b.keys(),b.values,is_convert=True)
bar
```

# 水平交错条形图

```python
a=ts.get_gdp_quarter()
b=a.iloc[:3,[3,5,7]]
b
```

```python
bar=pe.Bar("最近三个季度的GDP增加额",'水平交错条形图')
for i in range(len(b)):
    bar.add(str(a.iloc[i,0]),b.columns,b.iloc[i,:])
bar
```

# 垂直堆叠条形图

```python
b=a.iloc[:20,[3,5,7]]
b=b[::-1]
b=b.reset_index(drop=True)
b.head()
```

```python
bar=pe.Bar(title='三产业趋势条形图',subtitle="垂直堆叠条形图")
for i in range(b.shape[1]):
    bar.add(b.columns[i],b.index,b.iloc[:,i],is_stack=True)
bar
```

# 饼图

```python
a=ts.get_gdp_contrib()
a.head()
```

返回值说明：
* year :统计年度
* gdp_yoy :国内生产总值
* pi :第一产业献率(%)
* si :第二产业献率(%)
* industry:其中工业献率(%)
* ti :第三产业献率(%)

```python
b=a.iloc[0,[2,3,5]]
b
```

```python
pie=pe.Pie(title='三产业贡献率',subtitle='饼图')
pie.add(name='三产业',attr=b.keys(),value=b.values,is_label_show=True)
pie
```

# 箱线图
箱线图一般用来展现数据的分布（如上下四分位值、中位数等），同时，也可以用箱线图来反映数据的异常情况。

```python
a=ts.get_debtpaying_data(2018,1)
a.head()
```

* code,代码
* name,名称
* currentratio,流动比率
* quickratio,速动比率
* cashratio,现金比率
* icratio,利息支付倍数
* sheqratio,股东权益比率
* adratio,股东权益增长率

```python
any(a.sheqratio.isnull())
```

```python
boxplot=pe.Boxplot(title='A股股东权益比例',subtitle='箱线图')
boxplot.add(name='股东权益比例',x_axis=[a.sheqratio.name],y_axis=boxplot.prepare_data([a.sheqratio]))
boxplot
```

# 多箱线图¶
eg.A股每个地区的股东权益比例箱线图

```python
b=ts.get_stock_basics()
b.head()
```

* code,代码
* name,名称
* industry,所属行业
* area,地区
* pe,市盈率
* outstanding,流通股本(亿)
* totals,总股本(亿)
* totalAssets,总资产(万)
* liquidAssets,流动资产
* fixedAssets,固定资产
* reserved,公积金
* reservedPerShare,每股公积金
* esp,每股收益
* bvps,每股净资
* pb,市净率
* timeToMarket,上市日期
* undp,未分利润
* perundp, 每股未分配
* rev,收入同比(%)
* profit,利润同比(%)
* gpr,毛利率(%)
* npr,净利润率(%)
* holders,股东人数

```python
c=pd.merge(a.loc[:,['code','sheqratio']],b.loc[:,['area']],left_on='code',right_index=True,how='inner')
c.head()
```

```python
sheqratios=[]
labels=c.area.unique()
for i in labels:
    sheqratios.append(c.loc[c.area==i,'sheqratio'])
```

```python
boxplot=pe.Boxplot(title='各地域股东权益比例',subtitle='多箱线图')
boxplot.add(name='各地域',x_axis=labels,y_axis=boxplot.prepare_data(sheqratios))
boxplot
```

# 折线图
eg.股价折线图

```python
a=ts.get_k_data('000002',start='2018-06-06')
a.head()
```

```python
line=pe.Line(title='万科A股价图',subtitle='折线图')
line.add(
    name=a.open.name,
    x_axis=a.date,
    y_axis=a.open,
    yaxis_min='dataMin',
    yaxis_max='dataMax',
    xaxis_name=a.date.name,
    yaxis_name=a.open.name,
    line_type='dotted',#'solid', 'dashed', 'dotted'
    line_color='#123456',#颜色可以是英文或者颜色编码
)
line
```

# 多元折线图

```python
line=pe.Line(title='万科A股价图',subtitle='多元折线图')
for i in range(1,5):
    line.add(
        name=a.columns[i],
        x_axis=a.date,
        y_axis=a.iloc[:,i],
        yaxis_min='datamin',
        yaxis_max='datamax',
        xaxis_name=a.date.name,
        yaxis_name='价格',
        
    )
line
```

# 散点图
eg.上市公司总股本和市盈率的关系

```python
a=ts.get_stock_basics()
a=a.loc[:,['pe','totals']]
a.head()
```

```python
scatter=pe.Scatter(title='市值和市盈率')
scatter.add(name='散点图',x_axis=a.pe,y_axis=a.totals)
scatter
```

# 分组散点图

```python
a=ts.get_stock_basics()
a.head()
```

```python
areas=a.area.unique()
scatter=pe.Scatter(title='各地区市盈率和市值图',subtitle='分组散点图',height=800)
for i in areas:
    scatter.add(
        name=i,
        x_axis=a.loc[a.area==i,:].pe,
        y_axis=a.loc[a.area==i,:].totals,
        legend_pos='90%',
        legend_orient='vertical',
    )
scatter
```

# 气泡图的绘制

```python
a=pd.DataFrame(pd.np.random.randn(50,3),columns=list('xyz'))
a.head()
```

```python
scatter=pe.Scatter('气泡图')
scatter.add(name='随机数',x_axis=a.x,y_axis=a.y,extra_data=a.z.values.tolist(),is_visualmap=True,visual_dimension=2,visual_type='size',
           visual_range=[min(a.z),max(a.z)])
scatter
```
