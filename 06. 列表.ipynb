# 列表
* Python有6个序列的内置类型，但最常见的是列表和元组。
* 序列都可以进行的操作包括索引，切片，加，乘，检查成员。
* Python已经内置确定序列的长度以及确定最大和最小的元素的方法。
* 列表是最常用的Python数据类型，它可以作为一个方括号内的逗号分隔值出现。
* 列表的数据项不需要具有相同的类型
* 创建一个列表，只要把逗号分隔的不同的数据项使用方括号括起来即可。

```python
list1=[1,2,"hello",True]
print(list1)
```

# 访问列表中的值
* 与字符串的索引一样，列表索引从0开始。列表可以进行截取、组合等。
* 使用下标索引来访问列表中的值，同样你也可以使用方括号的形式截取字符

```python
print(list1,list1[0],list1[0:3])
```

# 更新列表
* 你可以对列表的数据项进行修改或更新，你也可以使用append()方法来添加列表项

```python
print(list1)
list1[0]=111
print(list1)
```

# 删除列表元素
* 可以使用 del 语句来删除列表的的元素

```python
print(list1)
del list1[0]
print(list1)
```

# 列表脚本操作符
列表对 + 和 \* 的操作符与字符串相似。+ 号用于组合列表，* 号用于重复列表。

```python
list1=[1,2,"hello",True]
print(list1) # 打印list1
print(len(list1)) # 求list1列表的长度
list2=list1+list1 # 两个list拼接
print(list2)
list3=list1*3
print(list3)
print(1 in list1)
for x in list1:print(x,end='\n')
```

# 列表截取与拼接

```python
print(list1)
print(list1[0])
print(list1[-1])
print(list1[:-1]) # [)
```

```python
print(list1+list1)
```

# 嵌套列表

```python
print(list1)
list2=[list1,list1,list1*2]
print(list2)
```

```python
print(list2[0])
print(list2[0][1])
```

# 列表函数和方法

```python
list1=range(10)
print(list1)
print(len(list1))
print(max(list1))
print(min(list1))
tuple1=(1,2,3)
print(type(tuple1),type(list(tuple1)))
```

```python
list1=[1,2,3,4]
# append 在列表末尾添加新的对象
print(list1)
list1.append("world")
print(list1)
```

```python
# count 统计某个元素在列表中出现的次数
print(list1.count(1))
xa=[0,False]
print(1==True)
```

```python
# extend 在列表末尾一次性追加另一个序列中的多个值（用新列表扩展原来的列表）
list1=[1,2,3,4]
list2=[4,5,6]
list1.extend(list2)
print(list1)
```

```python
# index  从列表中找出某个值第一个匹配项的索引位置
print(list1)
print(list1.index(4))
```

```python
# insert 将对象插入列表
print(list1)
list1.insert(0,10)
print(list1)
```

```python
list1.insert(1,111)
print(list1)
```

```python
# pop 移除列表中的一个元素（默认最后一个元素），并且返回该元素的值
print(list1)
print(list1.pop())
print(list1)
```

```python
print(list1.pop(0))
print(list1)
```

```python
# remove 移除列表中某个值的第一个匹配项
print(list1)
list1.remove(10)
print(list1)
```

```python
# reverse 反向列表中元素
print(list1)
print(list1.reverse())
print(list1)
```

```python
# list.sort(cmp=None, key=None, reverse=False) 对原列表进行排序
print(list1)
list1.sort(reverse=False)
print(list1)
```

```python
# clear 清空列表
print(list1)
list1.clear()
print(list1)
```

```python
# copy 复制列表
list1=[1,2,2]
print(list1)
list2=list1.copy()
print(list2)
```

```python
list1=[1,2,2]
print(list1)
list2=list1.copy()
list2[0]=111
print(list1,list2)
```

```python
list1=[1,2,2]
print(list1)
list2=list1
list2[0]=111
print(list1,list2)
```
